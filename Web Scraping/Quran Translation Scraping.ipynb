```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import csv
```

```python
# Function to scrape Quran translations and save to CSV
def scrape_quran_translations(base_url, output_file):
    with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Surah Number", "Verse Number", "Verse Text"])

        # Iterate over surahs from 1 to 114
        for surah_number in range(1, 115):
            surah_url = f"{base_url}{surah_number:03d}.html"
            print("Extracting from:", surah_url)

            try:
                # Send an HTTP GET request to the URL
                response = requests.get(surah_url)
                response.raise_for_status()

                soup = BeautifulSoup(response.content, 'html.parser')
                quran_text_div = soup.find('div', {'id': 'quran-text'})
                verse_elements = quran_text_div.find_all('p')

                for verse_number, verse in enumerate(verse_elements, start=1):
                    verse_text = verse.get_text(strip=True)
                    writer.writerow([surah_number, verse_number, verse_text])
                    print(f"Surah {surah_number}, Verse {verse_number} saved successfully.")
            except requests.exceptions.RequestException as e:
                print(f"Error fetching website content from {surah_url}: {e}")

if __name__ == '__main__':
    base_url = "https://www.clearquran.com/"
    output_file = 'quran_translations.csv'
    
    scrape_quran_translations(base_url, output_file)
```

```python
pillar_keywords = {
    "Shahada": ["faith", "belief", "testimony"],
    "Salat": ["prayer", "worship", "Salat"],
    "Zakat": ["charity", "give", "poor"],
    "Sawm": ["fasting", "fast"],
    "Hajj": ["pilgrimage", "Hajj", "Makkah", "Kaaba"]
}

df=pd.read_csv('quran_translations.csv')
```

```python
# Function to categorize ayahs based on keywords
def categorize_ayah(ayah_text):
    for pillar, keywords in pillar_keywords.items():
        if any(keyword in ayah_text.lower() for keyword in keywords):
            return pillar
    return "Other"

# Apply the categorization function to each verse
df['Pillar'] = df['Verse Text'].apply(categorize_ayah)

# Save the categorized data to a new CSV file
df.to_csv('quran_categorized.csv', index=False)
```
